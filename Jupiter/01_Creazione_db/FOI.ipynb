```python
foi_df_ita=foi_df[(foi_df["Territorio"]=="Italia") & (foi_df["Misura"]=='numeri indici') &(foi_df["COICOP Rev. Istat"]=='indice generale')]
print(foi_df_ita)
```

```python
foi_df_ita2=foi_df_ita[["TIME", "Seleziona periodo" , "Value"]]
foi_df_ita2
```

per trimestralizzare la serie faccio la media dei 3 mesi precedenti

```python
foi_df_ita2=foi_df_ita2.sort_values(['TIME'])
foi_df_ita2
```

```python
foi_df_ita2['Value -1'] = foi_df_ita2['Value'].shift(1) 
foi_df_ita2['Value -2'] = foi_df_ita2['Value'].shift(2) 

foi_df_ita2
```

```python
foi_df_ita3=foi_df_ita2[pd.notnull(foi_df_ita2['Value -2'])]
foi_df_ita3['FOI'] =(foi_df_ita2['Value']+foi_df_ita2['Value -1']+foi_df_ita2['Value -2'])/3
foi_df_ita3
```

```python
#tengo solo i trimestri
foi_df_ita3["MESE"]=foi_df_ita3["TIME"].str[5:7]
foi_df_ita3
```

```python
# get data type of column 
foi_df_ita3.dtypes['MESE']
```

```python
foi_df_ita3["MESE"]=foi_df_ita3["MESE"].astype(float)
foi_df_ita3["QUARTER"]=foi_df_ita3["MESE"]/3
foi_df_ita3
```

```python
foi_df_ita4=foi_df_ita3[(foi_df_ita3["QUARTER"]==1)|(foi_df_ita3["QUARTER"]==2)|(foi_df_ita3["QUARTER"]==3)|(foi_df_ita3["QUARTER"]==4)]
```

```python
foi_df_ita4=foi_df_ita4[["TIME","FOI","QUARTER"]]
foi_df_ita4["YEAR"]=foi_df_ita4["TIME"].str[:4]
foi_df_ita4["YEAR"]=foi_df_ita4["YEAR"].astype(int)
foi_df_ita4["QUARTER"]=foi_df_ita4["QUARTER"].astype(int)
foi_df_ita4=foi_df_ita4[["TIME","YEAR","QUARTER","FOI"]]
foi_df_ita4
```
