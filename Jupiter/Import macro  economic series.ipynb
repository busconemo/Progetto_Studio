# GDP

```python
# Importazione delle librerie necessarie
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gdp = pd.read_excel (r'C:\Users\busco\OneDrive\Documenti\Python Scripts\Repo_tassi_dec\Progetto_Studio\Data\GDP.xlsx', sheet_name="GDP modif")
```

```python
gdp.head()
```

```python
gdp.tail()
```

```python
gdp_df=pd.DataFrame(gdp)
```

```python
gdp_df["Quarter"]=gdp_df["TIME"].str.strip().str[6]
```

```python
gdp_df.head()
```

Creo la variabile GDP come il rapporto annuale del gdp a livelli

```python
gdp_df=gdp_df.sort_values(['TIME'])
```

```python
gdp_df['Previous GDP'] = gdp_df.sort_values('TIME').groupby('Quarter')['Italy'].shift(1) 
```

```python
gdp_df[gdp_df["Quarter"]=="1"]
```

```python
gdp_df["GDP"]=gdp_df["Italy"]/gdp_df["Previous GDP"] -1
```

```python
gdp_df
```

```python
def plot_series(time, series, format="-", start=0, end=None):
    plt.plot(time[start:end], series[start:end], format)
    plt.xlabel("Time")
    plt.ylabel("Value")
    plt.grid(True)
```

```python
#plot 
plt.figure(figsize=(30, 10))
plot_series(gdp_df["TIME"], gdp_df["GDP"],start=80)
```
